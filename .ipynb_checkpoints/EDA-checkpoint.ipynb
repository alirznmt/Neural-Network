```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data_dir = "/kaggle/working/fashionmnist"

# Define transformation
transform = transforms.Compose([
    transforms.ToTensor(),
])


# Create training ,test and  validation datasets
dataset = datasets.FashionMNIST(data_dir, download=True, train=True, transform=transform)
```

```python
class_counts = [0] * 10
for _, label in dataset:
    class_counts[label] += 1

plt.bar(range(10), class_counts)
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.title('Class Distribution')
plt.show()
```

```python
pixel_values = torch.cat([img.view(-1) for img, _ in dataset], dim=0)
mean_pixel = pixel_values.mean().item()
std_pixel = pixel_values.std().item()

print(f"Mean Pixel Value: {mean_pixel}")
print(f"Standard Deviation of Pixel Values: {std_pixel}")
```

```python
sample_loader = torch.utils.data.DataLoader(dataset, batch_size=5, shuffle=True)

def show_sample_images(images, labels):
    fig, axes = plt.subplots(1, len(images), figsize=(12, 3))
    for ax, img, label in zip(axes, images, labels):
        ax.imshow(img.view(28, 28).cpu().numpy(), cmap='gray')  # Add .cpu().numpy()
        ax.set_title(f"Label: {label.item()}")  # Add .item() to extract the value from the tensor
        ax.axis('off')

for batch in sample_loader:
    images, labels = batch
    show_sample_images(images, labels)
    break  # Display only the first batch

plt.show()
```

This visualization helps to understand how pixel values in the FashionMNIST dataset correlate with each other. Warmer (darker) colors indicate stronger positive correlations, while cooler (lighter) colors indicate stronger negative correlations.

```python
correlation_matrix = np.corrcoef(pixel_values.view(-1, 28*28).numpy(), rowvar=False)

sns.heatmap(correlation_matrix, cmap='coolwarm', xticklabels=False, yticklabels=False)
plt.title('Pixel Value Correlation Heatmap')
plt.show()
```

This function is useful for inspecting specific images from the dataset that are suspected to be anomalies. The anomalies are highlighted for closer examinatio

```python
def plot_images_with_anomalies(dataset, num_images=5):
    anomaly_indices = [5000, 8000, 12000, 15000, 18000]  # Replace with indices where anomalies are suspected

    for i in range(num_images):
        idx = anomaly_indices[i]
        img, label = dataset[idx]
        plt.subplot(1, num_images, i + 1)
        plt.imshow(img.view(28, 28), cmap='gray')
        plt.title(f"Label: {label}")
        plt.axis('off')
    plt.show()

plot_images_with_anomalies(dataset)
```

```python
from skimage import filters

def apply_edge_detection(image):
    edges = filters.sobel(image.numpy())
    return torch.tensor(edges, dtype=torch.float32)

# Example of applying edge detection to a sample image
sample_image = dataset[0][0].view(28, 28)
edge_image = apply_edge_detection(sample_image)

plt.subplot(1, 2, 1)
plt.imshow(sample_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edge_image, cmap='viridis')
plt.title('Edge Detection')
plt.axis('off')

plt.show()
```

```python
from torchvision import transforms

augmentation_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2)
])

augmented_dataset = datasets.FashionMNIST(root='./data', train=True, download=True, transform=augmentation_transform)

# Visualize original and augmented images
original_image, _ = dataset[0]
augmented_image, _ = augmented_dataset[0]

plt.subplot(1, 2, 1)
plt.imshow(original_image.view(28, 28), cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(augmented_image.permute(1, 2, 0))  # Change to permute
plt.title('Augmented Image')
plt.axis('off')

plt.show()
```
